```python
from scipy.stats import pointbiserialr

#What factors influence policy renewal rate?
#Correlation between the renewal status and customer satisfaction
plt.figure(figsize=(8, 6))
sns.boxplot(x="Renewal_Status", y="Customer_Satisfaction_Score", data=df)
plt.title("Renewal Status vs Customer Satisfaction Score")
plt.xlabel("Renewal Status")
plt.ylabel("Customer Satisfaction Score")
plt.show()

# Point-Biserial Correlation
renewal_status = df['Renewal_Status'].map({'Yes': 1, 'No': 0})  # Convert to binary
customer_satisfaction = df['Customer_Satisfaction_Score']

# Calculate correlation
corr, p_value = pointbiserialr(renewal_status, customer_satisfaction)
print(f"Point-Biserial Correlation: {corr:.3f}")
print(f"P-value: {p_value:.3f}")

# Interpretation
if p_value < 0.05:
    print("The correlation is statistically significant.")
else:
    print("The correlation is not statistically significant.")
```

```python
# Set figure size and style
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Create the countplot with custom colors and alpha
ax = sns.countplot(x="Discounts_Offered", 
                  hue="Renewal_Status", 
                  data=df,
                  alpha=0.8)

# Customize the title and labels with better formatting
plt.title("Impact of Discounts on Policy Renewal Status", 
          fontsize=14, 
          pad=20, 
          fontweight='bold')
plt.xlabel("Discounts Offered", fontsize=12)
plt.ylabel("Number of Policies", fontsize=12)

# Add value labels on top of each bar
for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=3)

# Adjust legend
plt.legend(title="Renewal Status", bbox_to_anchor=(1.05, 1), loc='upper left')

# Add percentage annotations
total = len(df)
for i in df['Discounts_Offered'].unique():
    subset = df[df['Discounts_Offered'] == i]
    renewal_rate = (len(subset[subset['Renewal_Status'] == 'Yes']) / len(subset)) * 100
    plt.text(df['Discounts_Offered'].unique().tolist().index(i), 
             ax.get_ylim()[1], 
             f'Renewal Rate: {renewal_rate:.1f}%', 
             ha='center',
             va='bottom')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()

# Print statistical summary
print("\nStatistical Summary:")
print("-" * 50)
renewal_by_discount = pd.crosstab(df['Discounts_Offered'], 
                                 df['Renewal_Status'], 
                                 normalize='index') * 100
print("\nRenewal Rates by Discount Category:")
print(renewal_by_discount.round(2))

# Chi-square test for independence
from scipy.stats import chi2_contingency
chi2, p_value, dof, expected = chi2_contingency(pd.crosstab(df['Discounts_Offered'], 
                                                           df['Renewal_Status']))
print(f"\nChi-square test p-value: {p_value:.4f}")
```

```python
#Which line of business has he highest renewal rate?
# Renewal rate by Policy Type
renewal_by_policy = df.groupby('Policy_Type')['Renewal_Status'].value_counts(normalize=True) * 100
print("Renewal Rate by Policy Type:")
print(renewal_by_policy)
```

```python
#What is the bestline of business for brokers and producers?
# Average Premium by Policy Type
avg_premium_by_policy = df.groupby('Policy_Type')['Premium'].mean()
print("Average Premium by Policy Type:")
print(avg_premium_by_policy)
```

```python
#How customer dempgraphics impact renewal?
# Renewal rate by Income Level
renewal_by_income = df.groupby('Income_Level')['Renewal_Status'].value_counts(normalize=True) * 100
print("Renewal Rate by Income Level:")
print(renewal_by_income)

# Renewal rate by Marital Status
renewal_by_marital = df.groupby('Marital_Status')['Renewal_Status'].value_counts(normalize=True) * 100
print("Renewal Rate by Marital Status:")
print(renewal_by_marital)
```
